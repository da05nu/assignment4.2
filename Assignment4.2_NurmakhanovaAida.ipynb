```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r"C:\Users\Пользователь\OneDrive\Desktop\datavisual\livestock2024.xlsx"
xls = pd.ExcelFile(file_path)
print(xls.sheet_names)
```

```python
df = pd.read_excel(file_path, sheet_name='1.', skiprows=5)

df.columns = [
    'Category',
    'Agricultural_2024', 'Agricultural_2023', 'Agricultural_%_to_2023',
    'Enterprises_2024', 'Enterprises_2023', 'Enterprises_%_to_2023',
    'Entrepreneurs_2024', 'Entrepreneurs_2023', 'Entrepreneurs_%_to_2023',
    'Households_2024', 'Households_2023', 'Households_%_to_2023',
    'All_2024', 'All_2023', 'All_%_to_2023'
]

df = df.dropna(subset=['Category'])
df.reset_index(drop=True, inplace=True) 
df.head(10) 
```

```python
df.info()
df.describe(include='all') 
```

```python
df.isnull().sum() 
```

```python
df_clean = df.dropna(subset=[df.columns[1], df.columns[2]])
df_clean.shape
```

```python
hist_data = df['All_2024'].dropna().head(10)

plt.figure(figsize=(10, 6))
plt.hist(hist_data, bins=5, edgecolor='black', alpha=0.7)
plt.title('Histogram of Total Livestock (Top 10 Categories) in 2024')
plt.xlabel('Number of Animals')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
bar_data = df[['Category', 'All_2024']].dropna().head(10)

plt.figure(figsize=(12, 7))
plt.barh(bar_data['Category'], bar_data['All_2024'], color='pink')
plt.title('Top 10 Livestock Categories by Total Count in 2024')
plt.xlabel('Number of Animals')
plt.ylabel('Livestock Category')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()


```
